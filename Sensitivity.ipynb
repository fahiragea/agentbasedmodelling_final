```python
import itertools
import random
import pandas as pd
from model import AdaptationModel
import Sensitivity

# Define parameter ranges
financial_range = [0.16, 0.2, 0.24]
trait_range = [0.16, 0.2, 0.24]
ext_range = [0.16, 0.2, 0.24]

# Store results
results = pd.DataFrame(columns=['run', 'financial_weight', 'trait_weight', 'ext_weight', 'agent_data', "model_data"])
run = 0
# Iterate over each combination of parameters
for financial, trait, ext in itertools.product(financial_range, trait_range, ext_range):
    results.loc[run] = Sensitivity.analyse(run, financial, trait, ext)
    run+=1

```

```python
end_res = pd.DataFrame(columns=['run', 'financial_weight', 'trait_weight', 'ext_weight', "#adapted",'adaption_time_avg', 'total_dmg'])
for result in results.iterrows():
    run = result[1][0]
    f = result[1][1]
    t = result[1][2]
    e = result[1][3]
    df_m = result[1][5]
    df_a =result[1][4]
    adapted = df_m.loc[99]
    data_99 = df_a.xs(99, level='Step')
    adaption_time_avg = data_99['adapted_time'].mean()
    total_dmg = df_a['FloodDamageActual'].sum()
    end_res.loc[run] = [run,f,t,e,adapted,adaption_time_avg, total_dmg]
```
